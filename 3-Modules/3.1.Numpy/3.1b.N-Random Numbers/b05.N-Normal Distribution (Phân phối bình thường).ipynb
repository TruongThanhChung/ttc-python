Normal Distribution
The Normal Distribution is one of the most important distributions.

It is also called the Gaussian Distribution after the German mathematician Carl Friedrich Gaus.

It fits the probability distribution of many events, eg. IQ Scores, Heartbeat tc.

Use the random.normal() method to get a Normal Data Distribtion.

It has three parmeters:

loc - (Mean) where the peak of the bel exists.

scale - (Standard Deviation) how flat the graph distributio should be.

size - The shape of the 
Phân phối bình thường
Phân phối chuẩn là một trong những phân phối quan trọng nhất.

Nó còn được gọi là Phân phối Gaussian theo tên nhà toán học người Đức Carl Friedrich Gaus.

Nó phù hợp với phân bố xác suất của nhiều sự kiện, vd. Điểm IQ, nhịp tim, .v.

Sử dụng random.n ormal()phương pháp này để có được Phân phối dữ liệu bình tường.

Nó có ba ham số:

loc- (Nghĩa là) nơi có đỉh chuông.

scale- (Độ lệch chuẩn) mức độ nphân bố đồ thị phải phẳng hư thế nào.

size- Hình dạng của mảng trả về.returned array.

```python
# Generate a random normal distribution of size 2x3:
# Tạo phân phối chuẩn ngẫu nhiên có kích thước 2x3:
from numpy import random

x = random.normal(size=(2, 3))

print(x)
```

```python
# Generate a random normal distribution of size 2x3 with mean at 1 and standard deviation of 2:
# Tạo phân phối chuẩn ngẫu nhiên có kích thước 2x3 với giá trị trung bình là 1 và độ lệch chuẩn là 2:
from numpy import random

x = random.normal(loc=1, scale=2, size=(2, 3))

print(x)
```

Visualization of Normal Distribution
Trực quan hóa phân phối chuẩn

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(size=1000), hist=False)

plt.show()
```

Note: The curve of a Normal Distribution is also known as the Bell Curve because of the bell-shaped curve.
