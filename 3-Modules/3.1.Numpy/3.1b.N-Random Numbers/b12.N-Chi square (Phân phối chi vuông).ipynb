Chi Square Distribution
Chi Square distribution is used as a basis to verify the hypothesis.

It has two parameter:

df - (degree of freedm).

size - The shape of the returned
Phân phối chi vuông
Phân phối Chi Square được sử dụng làm cơ sở để kiểm chứng giả thuyết.

Nó có hai tham s:

df- (mức độ tự o).

size- Hình dạng của mảng trả về. array.

```python
# Draw out a sample for chi squared distribution with degree of freedom 2 with size 2x3:
# Vẽ mẫu phân bố chi bình phương bậc tự do 2 với kích thước 2x3:
from numpy import random

x = random.chisquare(df=2, size=(2, 3))

print(x)
```

Visualization of Chi Square Distribution
Trực quan hóa phân phối Chi Square

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.chisquare(df=1, size=1000), hist=False)

plt.show()
```
