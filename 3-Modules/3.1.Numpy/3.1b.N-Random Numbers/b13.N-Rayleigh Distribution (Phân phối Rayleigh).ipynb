Rayleigh Distribution
Rayleigh distribution is used in signal processing.

It has two parameter:

scale - (standard deviation) decides how flat the distribution will be default 10).

size - The shape of the returned
Phân phối Rayleigh
Phân phối Rayleigh được sử dụng trong xử lý tín hiệu.

Nó có hai tham s:

scale- (độ lệch chuẩn) quyết định mức độ phẳng của phân phối sẽ được mặc định là 10).

size- Hình dạng của mảng trả về. array.

```python
# Draw out a sample for rayleigh distribution with scale of 2 with size 2x3:
# Vẽ ra mẫu phân bố rayleigh với tỷ lệ 2 với kích thước 2x3:
from numpy import random

x = random.rayleigh(scale=2, size=(2, 3))

print(x)
```

Visualization of Rayleigh Distribution
Trực quan hóa phân phối Rayleigh

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.rayleigh(size=1000), hist=False)

plt.show()
```

Similarity Between Rayleigh and Chi Square Distribution
At unit stddev and 2 degrees of freedom rayleigh and chi square represent the same distributions.
Điểm tương đồng giữa phân bố Rayleigh và Chi Square
Tại đơn vị stddev và 2 bậc tự do rayleigh và chi bình phương thể hiện sự phân bố giống nhau.
